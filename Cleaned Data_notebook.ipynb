## Convert the numeric columns from Object to numeric
### For Converting the data into numeric first clean the columnswhich have (, |)

```python
columns = ["rating_count",'rating']
for col in columns:
    data_set[col] = data_set[col].str.replace(r'[,]','',regex=True)
    data_set[col] = data_set[col].str.replace(r'[|]','None',regex=True)
```

### checking the clean data

```python
data_set["rating_count"]
```

### Droping the null value rows from the Data

```python
data_set1 = data_set.dropna(subset=['rating_count',"rating"])
```

# converting the data type from Object to numeric 

```python
data_set1["rating"] = pd.to_numeric(data_set1["rating"],errors='coerce')
```

### converting the whole data

```python
data_set1 = data_set1.astype({'rating_count':int,"rating":float})
```

### check the null values

```python
data_set1.isnull().sum()
```

### rating	rating_count these columns converted int, float

### verifying Data

### cleaning data and converting data

```python
columns = ['discounted_price','actual_price',"discount_percentage"]
for col in columns:
    data_set1[col] = data_set1[col].str.replace(r'[%₹,'']','',regex=True)
```

```python
data_set1=data_set1.astype({"discount_percentage":float,'actual_price':float,'discounted_price':float})
```

### Creating a product Table for products

```python
Product_table = data_set1[['product_id', 'product_name', 'category', 'discounted_price','actual_price', 'discount_percentage', 'rating', 'rating_count','about_product']]
```

### Ensuring data Category Names Minimal

```python
Product_table['category'] = Product_table['category'].str.split('|').str[0]
```

### Making columns names in Camel Case Format
